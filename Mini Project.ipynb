```python
import requests
```

```python
url = 'https://www.my11circle.com/fantasy-sports/cricket-records/most-runs-in-odi/'
```

```python
data = requests.get(url)
```

```python
BeautifulSoup(data.text,'html')
```

```python
Soup = BeautifulSoup(data.text,'html')
```

```python
Soup.find('h1')
```

```python
Soup.find_all('h4')
```

```python
Soup.find_all(['h1','h2','h3'])
```

```python
Soup.find_all('p')
```

```python
para = Soup.find_all('p')
```

```python
for pa in para:
  print(pa.get_text().strip())
```

```python
Soup.find_all('table')[0]
```

```python
cri = Soup.find_all('table')[0]
```

```python
Soup.find_all('th') 
```

```python
ball = cri.find_all('th')
```

```python
ball
```

```python
cricdata = [title.text.strip() for title in ball]
```

```python
cricdata
```

```python
pip install pandas
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = cricdata)
```

```python
df
```

```python
columns = cri.find_all('tr')
```

```python
columns
```

```python
for out in columns[1:]:
    crick = out.find_all('td')
    cricket = [data.text.strip() for data in crick]
    length = len(df)
    df.loc[length] = cricket
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.to_csv(r'D:\web scraping\cricketdatas.csv',index = False)
```

```python
import seaborn as sn
import matplotlib.pyplot as pt
sn.set(color_codes = True)
```

```python
file_path = r'D:\web scraping\cricketdata.csv'
cd = pd.read_csv(file_path)
print(cd)
```

```python
cd.head(5)
```

```python
cd.info()
```

```python
sn.lineplot(x='Mat',y='Runs',data=cd)
```

```python
pt.figure(figsize=(20,10))
sn.stripplot(x='Player',y='Runs',data=cd)
pt.title('cricket data',fontsize=30)
pt.xlabel('Player',fontsize=33)
pt.ylabel('Runs',fontsize=33)
pt.xticks(rotation=90,fontsize=20)
pt.yticks(rotation=90,fontsize=20)
pt.show()
```

```python
pt.figure(figsize=(20,10))
sn.violinplot(x='Player',y='Runs',data=cd)
pt.title('cricket data',fontsize=30)
pt.xlabel('Player',fontsize=33)
pt.ylabel('Runs',fontsize=33)
pt.xticks(rotation=90,fontsize=20)
pt.yticks(rotation=90,fontsize=20)
pt.show()
```

```python
pt.figure(figsize=(20,10))
sn.scatterplot(x='Mat',y='Player',data=cd)
pt.title('cricket data',fontsize=30)
pt.xlabel('Mat',fontsize=33)
pt.ylabel('Player',fontsize=33)
pt.xticks(rotation=90,fontsize=20)
pt.yticks(rotation=90,fontsize=5)
pt.show()
```

```python
pt.figure(figsize=(20,10))
sn.histplot(x='Player',y='Runs',data=cd,color='green')
pt.title('cricket data',fontsize=30)
pt.xlabel('Player',fontsize=33)
pt.ylabel('Runs',fontsize=33)
pt.xticks(rotation=90,fontsize=20)
pt.yticks(rotation=90,fontsize=20)
pt.show()
```

```python
sn.barplot(x='Mat',y='Runs',data=cd,color='black')
```

```python
sn.rugplot(x='Mat',y='Avg',data=cd,color='red')
```

```python
pt.figure(figsize=(20,10))
sn.jointplot(x='Player',y='Runs',data=cd,color='blue')
pt.xlabel('Player',fontsize=20,)
pt.ylabel('Runs',fontsize=20)
pt.xticks(rotation=90,fontsize=10)
pt.yticks(rotation=90,fontsize=10)
pt.show()
```
